```python
import pandas as pd
import numpy as np

# dataset file path
file_path = r"C:\Users\user\OneDrive\Desktop\ZAALIMA PROJECT\CUSTOMER CHURN PREDICTION LTV ENGINE\data\raw\Telco-Customer-Churn.csv"

# loading dataset into dataframe
filepath = pd.read_csv(file_path)
```

```python
# DAY 4
# Checking Hidden Blank Values in Object Columns
#This step identifies hidden empty spaces present inside categorical/object columns of the dataset.

for col in filepath.select_dtypes(include="object").columns:
    print(f"\nColumn: {col}")
    print(filepath[col].isin([" "]).sum())
```

```python
# replacing blank spaces with NaN
#Blank space values are replaced with NaN to handle missing data properly during analysis.

filepath = filepath.replace(" ", pd.NA)
filepath.isnull().sum()
```

```python
# Datatype Inspection
# df.dtypes is used to inspect the datatype of each column in the dataset.

filepath.dtypes
```

```python
filepath["TotalCharges"] = pd.to_numeric(
    filepath["TotalCharges"],
    errors="coerce"
)

filepath.dtypes
```

```python
# Numerical Feature Statistics
#df.describe() generates statistical summaries like mean, median, minimum, maximum, and standard deviation for numerical columns.

filepath.describe()
```

```python
# Check Unique Values
#.nunique() is used to count the number of unique values present in each column.
for col in filepath.columns:
    print(f"\n{col}")
    print(filepath[col].unique())
```

```python
# Analyze Churn Distribution
#value_counts() is used to analyze the distribution of churned and non-churned customers in the dataset.
filepath["Churn"].value_counts(normalize=True) * 100
```
