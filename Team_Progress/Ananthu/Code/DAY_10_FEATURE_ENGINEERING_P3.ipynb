```python
import pandas as pd
import numpy as np

df = pd.read_csv(
    r"C:\Users\user\OneDrive\Desktop\ZAALIMA PROJECT\CUSTOMER CHURN PREDICTION LTV ENGINE\data\processed\telco_churn_day10_feature_engineered_v1.csv"
)
```

```python
df["RevenuePerMonth"] = (
    df["TotalCharges"] /
    (df["tenure"] + 1)
)
```

```python
df["CustomerAgeGroup"] = pd.cut(
    df["tenure"],
    bins=[0,12,24,48,72],
    labels=[
        "New",
        "Growing",
        "Established",
        "Loyal"
    ]
)
```

```python
conditions = [
    (df["tenure"] < 12),
    (df["tenure"] >= 12) &
    (df["tenure"] < 36),
    (df["tenure"] >= 36)
]

choices = [
    "High Risk",
    "Medium Risk",
    "Low Risk"
]

df["RiskSegment"] = np.select(
    conditions,
    choices,
    default="Medium Risk"
)
```

```python
df["MonthlyChargeCategory"] = pd.qcut(
    df["MonthlyCharges"],
    q=4,
    labels=[
        "Low",
        "Medium",
        "High",
        "Premium"
    ]
)
```

```python
contract_score = {
    "Month-to-month": 1,
    "One year": 2,
    "Two year": 3
}

df["ContractSecurityScore"] = (
    df["Contract"]
    .map(contract_score)
)
```

```python
service_cols = [
    "PhoneService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies"
]

df["TotalServices"] = (
    df[service_cols]
    .eq("Yes")
    .sum(axis=1)
)
```

```python
numerical_features = df.select_dtypes(
    include=["int64","float64"]
)

corr_matrix = numerical_features.corr()

corr_matrix["Churn"].sort_values(
    ascending=False
)
```
