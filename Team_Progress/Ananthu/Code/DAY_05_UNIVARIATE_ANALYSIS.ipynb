```python
# DAY 05
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\user\OneDrive\Desktop\ZAALIMA PROJECT\CUSTOMER CHURN PREDICTION LTV ENGINE\data\processed\telco_churn_day4_cleaned.csv"
df = pd.read_csv(file_path)
```

```python
df.describe()
```

```python
df['tenure'].hist(bins=30)
plt.title("Tenure Distribution")
plt.xlabel("Tenure")
plt.ylabel("Frequency")
plt.show()
```

```python
numerical_cols = [
    'tenure',
    'MonthlyCharges',
    'TotalCharges'
]
```

```python
numerical_cols = [
    'tenure',
    'MonthlyCharges',
    'TotalCharges'
]

df[numerical_cols].hist(
    figsize=(12,5),
    bins=30
)

plt.tight_layout()
plt.show()
```

```python

for col in numerical_cols:
    
    sns.kdeplot(df[col], fill=True)
    
    plt.title(f"KDE Plot - {col}")
    plt.show()
```

```python
for col in numerical_cols:
    
    sns.boxplot(x=df[col])
    
    plt.title(f"Boxplot - {col}")
    plt.show()
```

```python
df[numerical_cols].skew()
```

```python
sns.countplot(x='Churn', data=df)

plt.title("Churn Distribution")
plt.show()
```

```python
categorical_cols = [
    'gender',
    'Contract',
    'InternetService',
    'PaymentMethod'
]

for col in categorical_cols:
    
    plt.figure(figsize=(8,4))
    
    sns.countplot(x=col, data=df)
    
    plt.xticks(rotation=45)
    
    plt.title(f"Countplot - {col}")
    
    plt.show()
```

```python
df['Churn'].value_counts()
```

```python
df['Churn'].value_counts(normalize=True)
```

```python
df['Churn'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title("Churn Percentage Distribution")

plt.ylabel("")

plt.show()
```

# DAY 5 — Key Business Observations

1. Customer tenure distribution appears right-skewed, indicating many newer customers.

2. MonthlyCharges shows wide variability, suggesting multiple pricing plans.

3. TotalCharges contains high-value customers with long subscription durations.

4. Churn distribution is imbalanced, with non-churn customers dominating the dataset.

5. Certain categorical features like Contract type may strongly influence churn behavior.

6. Outliers are visible in TotalCharges, but may represent valuable long-term customers rather than errors.
