```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
```

```python
file_path = r"C:\Users\user\OneDrive\Desktop\ZAALIMA PROJECT\CUSTOMER CHURN PREDICTION LTV ENGINE\data\processed\telco_churn_day6_analysis.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
df.info()
df.shape
```

```python
df.isnull().sum()
```

```python
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)
```

```python
df.isnull().sum()
```

```python
numerical_cols = [
    "tenure",
    "MonthlyCharges",
    "TotalCharges"
]

for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
categorical_cols = df.select_dtypes(include=["object", "string"]).columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.drop(
    columns=["customerID"],
    inplace=True
)
```

```python
le = LabelEncoder()

df["Churn"] = le.fit_transform(df["Churn"])
```

```python
df["TenureGroup"] = pd.cut(
    df["tenure"],
    bins=[0,12,24,48,60,72],
    labels=[
        "0-12 Months",
        "12-24 Months",
        "24-48 Months",
        "48-60 Months",
        "60-72 Months"
    ]
)
```

```python
df["ChargeSegment"] = pd.cut(
    df["MonthlyCharges"],
    bins=[0,35,70,120],
    labels=[
        "Low",
        "Medium",
        "High"
    ]
)
```

```python
df.select_dtypes(include=["object", "string"]).columns
```

```python
df = pd.get_dummies(
    df,
    drop_first=True
)
```
