```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../data/processed/day10_feature_engineering.csv")
```

```python
df.head()
```

```python
df['Tenure_Group'] = pd.cut(
    df['tenure'],
    bins=[0, 12, 24, 48, 72],
    labels=['New', 'Regular', 'Loyal', 'Very Loyal']
)
```

```python
df['Tenure_Group'].value_counts()
```

```python
df['Charge_Segment'] = pd.cut(
    df['MonthlyCharges'],
    bins=[0, 35, 70, 120],
    labels=['Low', 'Medium', 'High']
)
```

```python
df['Charge_Segment'].value_counts()
```

```python
df[['tenure',
    'Tenure_Group',
    'MonthlyCharges',
    'Charge_Segment']].head(10)
```

```python
df.to_csv(
    "../../data/processed/day11_customer_segmentation.csv",
    index=False
)
```
