```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../data/processed/ltv_dataset.csv")
```

```python
df[['Estimated_LTV']].head()
```

```python
df['LTV_Group'] = pd.qcut(
    df['Estimated_LTV'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
df['LTV_Group'].value_counts()
```

```python
df[['Estimated_LTV', 'LTV_Group']].head(10)
```

```python
sns.countplot(x='LTV_Group', data=df)
plt.title('Customer LTV Segments')
plt.show()
```

```python
df.groupby('LTV_Group')['Estimated_LTV'].mean()
```

```python
df.to_csv(
    "../../data/processed/ltv_segmented_dataset.csv",
    index=False
)
```

## Observation

- Customers were divided into Low, Medium, and High LTV segments.
- High LTV customers generated the highest revenue.
- Segmentation helps identify valuable customers.
- LTV grouping can support targeted retention strategies.
