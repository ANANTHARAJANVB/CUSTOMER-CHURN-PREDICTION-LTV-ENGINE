```python
from fastapi import FastAPI
```

```python
app = FastAPI()
```

```python
@app.get("/")
def home():
    return {"message": "Customer Churn API Running"}
```

```python
home()
```

```python
@app.get("/predict")
def predict_customer():
    return {
        "Customer_ID": 1001,
        "Prediction": "No Churn",
        "Risk_Score": 0.20
    }
```

```python
predict_customer()
```

```python
@app.get("/predict_highrisk")
def predict_highrisk():
    return {
        "Customer_ID": 1002,
        "Prediction": "Churn",
        "Risk_Score": 0.85
    }
```

```python
predict_highrisk()
```
