```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../data/processed/day11_customer_segmentation.csv")
```

```python
df.head()
```

```python
df['Contract_Risk'] = df['Contract'].map({
    'Month-to-month': 3,
    'One year': 2,
    'Two year': 1
})
```

```python
df[['Contract','Contract_Risk']].head(10)
```

```python
df['TotalCharges'] = pd.to_numeric(
    df['TotalCharges'],
    errors='coerce'
)
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(
    df['TotalCharges'].median()
)
```

```python
df['Revenue_Per_Tenure'] = (
    df['TotalCharges'] /
    (df['tenure'] + 1)
)
```

```python
df[['TotalCharges',
    'tenure',
    'Revenue_Per_Tenure']].head(10)
```

```python
df[['Tenure_Group',
    'Charge_Segment',
    'Contract_Risk',
    'Revenue_Per_Tenure']].head()
```

```python
df.to_csv(
    "../../data/processed/feature_engineered_telco.csv",
    index=False
)
```
