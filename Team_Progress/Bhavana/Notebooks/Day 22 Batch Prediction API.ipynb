```python
from fastapi import FastAPI
```

```python
app = FastAPI()
```

```python
@app.get("/batch_predict")
def batch_predict():
    return {
        "Total_Customers": 3,
        "Predictions": [
            {"Customer_ID": 1001, "Prediction": "No Churn"},
            {"Customer_ID": 1002, "Prediction": "Churn"},
            {"Customer_ID": 1003, "Prediction": "No Churn"}
        ]
    }
```

```python
batch_predict()
```

```python
@app.get("/summary")
def summary():
    return {
        "Total_Customers": 3,
        "Churn_Customers": 1,
        "Non_Churn_Customers": 2
    }
```

```python
summary()
```
