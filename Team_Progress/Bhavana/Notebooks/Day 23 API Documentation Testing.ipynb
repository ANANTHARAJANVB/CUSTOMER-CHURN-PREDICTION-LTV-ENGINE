```python
from fastapi import FastAPI
```

```python
app = FastAPI()
```

```python
@app.get("/")
def home():
    return {"message": "Customer Churn API Running"}

@app.get("/predict")
def predict_customer():
    return {
        "Customer_ID": 1001,
        "Prediction": "No Churn",
        "Risk_Score": 0.20
    }

@app.get("/batch_predict")
def batch_predict():
    return {
        "Total_Customers": 3,
        "Predictions": [
            {"Customer_ID": 1001, "Prediction": "No Churn"},
            {"Customer_ID": 1002, "Prediction": "Churn"},
            {"Customer_ID": 1003, "Prediction": "No Churn"}
        ]
    }
```

```python
home()
```

```python
predict_customer()
```

```python
batch_predict()
```

# API Documentation

| Endpoint | Purpose |
|----------|----------|
| / | Home API |
| /predict | Single Customer Prediction |
| /batch_predict | Batch Customer Prediction |
