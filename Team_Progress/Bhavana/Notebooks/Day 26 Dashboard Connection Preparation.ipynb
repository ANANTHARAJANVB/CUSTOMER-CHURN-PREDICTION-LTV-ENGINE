```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine(
    "postgresql://postgres:MyNewPassword123@localhost:5432/telco_churn_db"
)
```

```python
df = pd.read_sql(
    "SELECT * FROM telco_customers",
    engine
)
```

```python
df.shape
```

```python
df.columns
```

```python
total_customers = len(df)

churned_customers = len(df[df['Churn'] == 'Yes'])

active_customers = len(df[df['Churn'] == 'No'])

churn_percentage = round(
    (churned_customers / total_customers) * 100,
    2
)

print("Total Customers:", total_customers)
print("Churned Customers:", churned_customers)
print("Active Customers:", active_customers)
print("Churn Percentage:", churn_percentage)
```
