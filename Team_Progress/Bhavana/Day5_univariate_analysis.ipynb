```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['tenure'], bins=20)
plt.title("Tenure Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['MonthlyCharges'], bins=20)
plt.title("Monthly Charges Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['tenure'], fill=True)
plt.title("Tenure KDE Plot")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['MonthlyCharges'], fill=True)
plt.title("Monthly Charges KDE Plot")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['tenure'])
plt.title("Tenure Boxplot")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['MonthlyCharges'])
plt.title("Monthly Charges Boxplot")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df)
plt.title("Churn Count")
plt.show()
```

```python
df['Churn'].value_counts(normalize=True) * 100
```

```python
df['Churn'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title("Churn Distribution")
plt.ylabel("")
plt.show()
```

```python
df[['tenure','MonthlyCharges']].skew()
```
