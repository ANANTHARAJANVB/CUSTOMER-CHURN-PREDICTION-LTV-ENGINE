```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")
print(df.shape)
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['tenure'], bins=30, kde=True, color='steelblue')
plt.title('Tenure Distribution')
plt.savefig('../../docs/reports/eda_visualizations/tenure_distribution.png')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x='Contract', y='tenure', data=df, palette='muted')
plt.title('Contract Type vs Tenure')
plt.savefig('../../docs/reports/eda_visualizations/contract_vs_tenure_violin.png')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Contract', hue='Churn', data=df, palette='Set2')
plt.title('Contract Type vs Churn')
plt.savefig('../../docs/reports/eda_visualizations/contract_vs_churn.png')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Churn', y='tenure', data=df, palette='pastel')
plt.title('Tenure vs Churn')
plt.savefig('../../docs/reports/eda_visualizations/tenure_vs_churn_box.png')
plt.show()
```

## Key Findings
- Month-to-month customers churn the most
- Longer tenure customers are more stable
- Two-year contract customers rarely churn
